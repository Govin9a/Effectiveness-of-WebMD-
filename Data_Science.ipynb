```python
import pandas as pd

# create an empty list to store the extracted values
data = []

# iterate through each soup object in soup_list
for soup in soup_list:
    # find all the font-normal spans
    font_normal_spans = soup.find_all('span', class_='font-normal')
    # extract the text from each span and append to data list
    values = [span.text.strip() for span in font_normal_spans]
    data.append(values)

# create a dataframe from the extracted values with appropriate column names
df2 = pd.DataFrame(data, columns = [    'effectiveness_5', 'effectiveness_4', 'effectiveness_3', 'effectiveness_2', 'effectiveness_1',    'ease_of_use_5', 'ease_of_use_4', 'ease_of_use_3', 'ease_of_use_2', 'ease_of_use_1',    'satisfaction_5', 'satisfaction_4', 'satisfaction_3', 'satisfaction_2', 'satisfaction_1'])

# display the dataframe
df2
```

```python
rating_list = []

for soup in soup_list:
    rating_span = soup.find('span', class_='rat-num')
    if rating_span:
        rating = rating_span.text.strip()
        rating_list.append(rating)
    else:
        rating_list.append(None)

df_rating = pd.DataFrame(rating_list, columns=['Overall rating'])
df_rating
```

```python
conditions_df = pd.DataFrame(columns=['condition'])
for soup in soup_list:
    conditions = []
    for condition in soup.find_all('strong', {'class': 'condition'}):
        conditions.append(condition.text.split(': ')[1])
    for c in conditions:
        conditions_df = conditions_df.append({'condition': c}, ignore_index=True)
conditions_df
```

```python
import pandas as pd

details_data = []

for soup in soup_list:
    details_list = soup.find_all('div', {'class': 'details'})
    for details in details_list:
        spans = details.find_all('span')
        if len(spans) >= 4:
            name = spans[0].text.strip().replace('|', '')
            age_range = spans[1].text.strip().replace('|', '')
            gender = spans[2].text.strip().replace('|', '')
            duration = details.contents[3].strip().replace('|', '')
            patient_type = spans[3].text.strip().replace('|', '')
        else:
            name = 'Anonymous'
            age_range = ''
            gender = ''
            duration = details.contents[3].strip().replace('|', '')
            patient_type = 'Patient'
        details_data.append([name, age_range, gender, duration, patient_type])

df3 = pd.DataFrame(details_data, columns=['Name', 'Age Range', 'Gender', 'Duration', 'Patient Type'])
df3
```

```python
ratin = []
for soup in soup_list:
    for rating in soup.find_all('div', {'class': 'overall-rating'}):
        ratin.append(float(rating.find('strong').text.split()[2]))
df4 = pd.DataFrame({'Individual rating': ratin})
df4
```

```python
drug_names = []
for soup in soup_list:
    drug_name = soup.find('h1', {'class': 'drug-name'}).find('a').text.strip()
    drug_names.append(drug_name)

# create a pandas dataframe from the drug names
df5 = pd.DataFrame({'Drug Name': drug_names})

# print the dataframe
df5
```

```python
description_texts = []
for soup in soup_list:
    description_tags = soup.find_all('p', class_='description-text')
    for tag in description_tags:
        description_texts.append(tag.text.strip())
description_texts
dt = pd.DataFrame({'Comments': description_texts})
dt
```

```python
df = pd.concat([df_rating, df2, df3, conditions_df, df4, dt, df5], axis=1)
df
```
